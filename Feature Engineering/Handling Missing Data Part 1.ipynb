```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/data_science_job.csv')
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
```

```python
cols
```

```python
df[cols].sample(5)
```

```python
new_df = df[cols].dropna()
```

```python
for col in cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(new_df[col], kde=True)  # kde=True adds a Kernel Density Estimate curve
    plt.title(f'Histogram for {col}')
    plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(12, 5))

sns.histplot(df['training_hours'], bins=50, ax=ax, kde=True, color='orange')

sns.histplot(new_df['training_hours'], bins=50, ax=ax, kde=True, color='teal')

plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(12, 5))

sns.histplot(df['city_development_index'], bins=50, ax=ax, kde=True, color='orange')

sns.histplot(new_df['city_development_index'], bins=50, ax=ax, kde=True, color='teal')

plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(12, 5))

sns.histplot(df['experience'], bins=50, ax=ax, kde=True, color='orange')

sns.histplot(new_df['experience'], bins=50, ax=ax, kde=True, color='teal')

plt.show()
```

```python
temp = pd.concat([
            # percentage of observations per category, original data
            df['enrolled_university'].value_counts() / len(df),

            # percentage of observations per category, cca data
            new_df['enrolled_university'].value_counts() / len(new_df)
        ],
        axis=1)

# add column names
temp.columns = ['original', 'cca']

temp
```
