```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette("Spectral")

# Set a base style
sns.set_style("whitegrid")

# Customize specific style parameters
custom_params = {
    "axes.spines.right": False,
    "axes.spines.top": False,
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.color": "#cccccc",
    "axes.facecolor": "#f0f0f0",
    "axes.labelsize": 14,
    "axes.titlesize": 16,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12
}
sns.set_context("notebook", rc=custom_params)
```

```python
df = pd.read_csv('Datasets/weight-height.csv')
```

```python
df.head()
```

```python
df['Height'].describe()
```

```python
sns.histplot(df['Height'], kde=True, color='teal')
plt.show()
```

```python
sns.boxplot(x=df['Height'], color='orange')
plt.show()
```

```python
upper_limit = df['Height'].quantile(0.99)
upper_limit
```

```python
lower_limit = df['Height'].quantile(0.01)
lower_limit
```

```python
new_df = df.loc[(df['Height'] < upper_limit) & (df['Height'] > lower_limit)]
```

```python
sns.histplot(new_df['Height'], kde=True, color='teal')
plt.show()
```

```python
sns.boxplot(x=new_df['Height'], color='orange')
plt.show()
```

```python
df['Height'] = np.where(df['Height'] >= upper_limit,
        upper_limit,
        np.where(df['Height'] <= lower_limit,
        lower_limit,
        df['Height']))
```

```python
sns.histplot(df['Height'], kde=True, color='teal')
plt.show()
```

```python
sns.boxplot(x=df['Height'], color='orange')
plt.show()
```
