```python
df = pd.read_csv('Datasets/Titanic-Dataset.csv')
```

### 1. What are the central tendencies of the data?

### 2. What is the distribution of the data?
- Questions: How is the data distributed? Is it normal, skewed, or uniform?
- When to Use: Use these to visualize and understand the shape and spread of the data.

```python
# Creates a histogram to visualize the frequency distribution
numeric_df = df.select_dtypes(include='number')
for column in numeric_df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], bins=30, kde=True, color='skyblue')
    plt.title('Histogram with Density Curve for ' + column, fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
# Plots a density curve to visualize the distribution
numeric_df = df.select_dtypes(include='number')
for column in numeric_df.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[column].dropna(), bw_adjust=0.5, label=column)
    plt.title(f'Density Plot of {column}', fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()
```

### 7. What are the statistical properties of the data, such as skewness and kurtosis?
- Questions: How skewed is the data, and how heavy or light are the tails compared to a normal distribution?
- When to Use: Use these to assess the asymmetry and tailedness of the data distribution.

```python
skewness = df['Age'].skew()
kurtosis = df['Age'].kurtosis()
```

```python
import scipy.stats as stats
# Create a figure with subplots
plt.figure(figsize=(12, 6))

# Plot the distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title(f'Distribution (Skewness: {skewness:.2f})', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the Q-Q plot to visualize skewness
plt.subplot(1, 2, 2)
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title(f'Q-Q Plot', fontsize=14)

# Add titles and labels
plt.tight_layout()
plt.show()
```

```python
# Generate a distribution plot
plt.figure(figsize=(12, 6))

# Plot the distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title(f'Distribution (Skewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f})', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a title for kurtosis
plt.subplot(1, 2, 2)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', label='Mean')
plt.title(f'Distribution with Mean Line', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

### How can continuous data be effectively categorized?
- Questions: How can the data be binned or discretized to create categorical features?
- When to Use: Use these methods to convert continuous data into categorical data for easier analysis or modeling.

```python
# Define bin edges
bins = [0, 15, 25, 35, 50]

# Bin the data
df['age_category'] = pd.cut(df['Age'], bins=bins, labels=['0-15', '15-25', '25-35', '35-50'])
# Create the box plot
plt.figure(figsize=(10, 6))
sns.barplot(x='age_category', y=df['Age'].index, data=df, palette="pastel")
plt.title('Bar Plot of Binned Data (using pd.cut)', fontsize=16)
plt.xlabel('Binned Intervals')
plt.ylabel('Values')

# Show the plot
plt.show()
```

### How do various metrics or features correlate with each other?
- Questions: What are the relationships between different metrics or features?
- When to Use: Use these to explore relationships and correlations between multiple features.

```python
numerical_df = df.select_dtypes(include='number')
corr = numerical_df.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
ax = sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='gray')

# Add titles and labels
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Features', fontsize=14)

# Show the plot
plt.show()
```

### What are the most common values in categorical data?
- Questions: What are the frequencies of each category in the data?
- When to Use: Use this to understand the distribution of categorical values.

```python
# Create a bar plot using Seaborn
sns.barplot(x=df['Pclass'].value_counts().index, 
            y=df['Pclass'].value_counts().values,
            palette='viridis')

plt.xlabel('Pclass')  # Label for x-axis
plt.ylabel('Count')   # Label for y-axis
plt.title('Distribution of Pclass')  # Title of the plot
plt.show()
```
