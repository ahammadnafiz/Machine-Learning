### Matplotlib .plot() Parameters and Notes

#### Basic Syntax
```python
ax.plot(x, y, **kwargs)
```
- `x`: Array-like or list of x-coordinates.
- `y`: Array-like or list of y-coordinates.
- `**kwargs`: Additional keyword arguments to control the appearance of the plot.

#### Common Parameters

#### Color
- `color` or `c`: Specifies the color of the line.
  - Example: `'red'`, `'blue'`, `'#FFDD44'`

#### Line Style
- `linestyle` or `ls`: Specifies the style of the line.
  - Example: `'-'` (solid), `'--'` (dashed), `'-.'` (dash-dot), `':'` (dotted)

#### Line Width
- `linewidth` or `lw`: Specifies the width of the line.
  - Example: `2.0`

#### Marker
- `marker`: Specifies the marker style.
  - Example: `'o'` (circle), `'s'` (square), `'^'` (triangle up)

#### Marker Size
- `markersize` or `ms`: Specifies the size of the marker.
  - Example: `8`

#### Marker Color
- `markerfacecolor` or `mfc`: Specifies the face color of the marker.
  - Example: `'green'`
- `markeredgecolor` or `mec`: Specifies the edge color of the marker.
  - Example: `'black'`

#### Marker Edge Width
- `markeredgewidth` or `mew`: Specifies the width of the marker edge.
  - Example: `1.5`

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle='-.', linewidth=1, marker='o', markersize=6, markerfacecolor='red', markeredgecolor='black')
plt.show()
````

# Matplotlib .hist() Parameters and Notes

## Basic Syntax
```python
plt.hist(x, bins=None, **kwargs)
```
- `x`: Array-like, the data to be histogrammed.
- `bins`: Integer or sequence, optional, the number of bins or bin edges.

## Common Parameters

### Bins
- `bins`: Defines the number of equal-width bins or a sequence of bin edges.
  - Example: `10` (number of bins), `[0, 1, 2, 3]` (bin edges)

### Range
- `range`: Tuple, the lower and upper range of the bins.
  - Example: `(0, 1)`

### Density
- `density`: If `True`, the first element of the return tuple will be the counts normalized to form a probability density.
  - Example: `True` or `False`

### Weights
- `weights`: Array-like, same shape as `x`. Each value in `x` only contributes its associated weight towards the bin count (instead of 1).
  - Example: `[1, 2, 3, 4]`

### Cumulative
- `cumulative`: If `True`, computes a cumulative histogram.
  - Example: `True` or `False`

### Color
- `color`: Specifies the color of the bars.
  - Example: `'blue'`, `'#FFDD44'`

### Edge Color
- `edgecolor`: Specifies the color of the bar edges.
  - Example: `'black'`

### Line Width
- `linewidth`: Specifies the width of the bar edges.
  - Example: `1.5`

### Orientation
- `orientation`: Specifies the orientation of the histogram.
  - Example: `'vertical'`, `'horizontal'`

### Alpha
- `alpha`: The alpha blending value, between 0 (transparent) and 1 (opaque).
  - Example: `0.75`

````python
data = np.random.randn(1000)

plt.hist(data, bins=30, range=(-3, 3), density=True, color='blue', edgecolor='black', linewidth=1.5, alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data, color = 'black', linestyle = '--', label='Default')
ax.plot(data, color='black', linestyle = '--', 
           drawstyle='steps-post', label='steps-post')
ax.legend()
plt.show()
````

## Ticks, Labels, and Legends

````python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
plt.show()
````

````python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                           rotation=30, fontsize=8)
ax.set_xlabel('Stages')
ax.set_title('My first matplotlib plot')
plt.show()
````

````python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum(), color='black', label='one')
ax.plot(np.random.standard_normal(1000).cumsum(), color='blue',
       linestyle ='--', label = 'two')
ax.plot(np.random.standard_normal(1000).cumsum(), color='red', linestyle='--',
       label='three')
ax.legend()
plt.show()
````

````python
from datetime import datetime
fig, ax = plt.subplots()
data = pd.read_csv("examples/spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]
spx.plot(ax=ax, color="black")
crisis_data = [
(datetime(2007, 10, 11), "Peak of bull market"),
(datetime(2008, 3, 12), "Bear Stearns Fails"),
(datetime(2008, 9, 15), "Lehman Bankruptcy")
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
    xytext=(date, spx.asof(date) + 225),
    arrowprops=dict(facecolor="black", headwidth=4, width=2,
    headlength=4),
    horizontalalignment="left", verticalalignment="top")
# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_title("Important dates in the 2008-2009 financial crisis")
plt.show()
````
