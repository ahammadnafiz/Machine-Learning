## Reading Text Files in Pieces

Before we look at a large file, we make the pandas display settings more compact:

```python
pd.options.display.max_rows = 10
```

```python
result = pd.read_csv('examples/ex6.csv')
```

```python
result
```

If you want to read only a small number of rows (avoiding reading the entire file),
 specify that with nrows:

```python
pd.read_csv('examples/ex6.csv', nrows=10)
```

 To read a file in pieces, specify a chunksize as a number of rows:

```python
chunker = pd.read_csv('examples/ex6.csv', chunksize=1000)
```

```python
test_chusnker = pd.read_csv('examples/ex6.csv')
```

```python
test_chusnker['key'].value_counts()
```

```python
type(chunker)
```

```python
tot = pd.Series([], dtype = 'int64')
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(), fill_value = 0)
tot = tot.sort_values(ascending=False)
tot
```

```python
tot[:10]
```

```python
header, values = lines[0], lines[1:]
data_dict = {
    h: v for h, v in zip(header, zip(*values))
}
data_dict
```

```python
test = [
    [1,2,3],
    [4, 5, 6],
    [7, 8, 9]
]
print(list(zip(*test)))
```

```python
f = open("examples/ex7.csv")
class MyDialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
reader = csv.reader(f, dialect = MyDialect)
for i in reader:
    print(i)
```

## JSON Data

```python
 obj = """
 {"name": "Wes",
 "cities_lived": ["Akron", "Nashville", "New York", "San Francisco"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 34, "hobbies": ["guitars", "soccer"]},
              {"name": "Katie", "age": 42, "hobbies": ["diving", "art"]}]
 }
 """
```

 To convert a JSON string to Python form, use json.loads:

```python
import json
result = json.loads(obj)
result
```

 json.dumps, on the other hand, converts a Python object back to JSON:

```python
asjson = json.dumps(result)
asjson
```

```python
pd.DataFrame(result['siblings'], columns=['name', 'age'])
```

The pandas.read_json can automatically convert JSON datasets in specific arrange
ments into a Series or DataFrame. For example:

```python
data = pd.read_json('examples/example.json')
data
```

```python
from lxml import objectify
```

```python
path = 'datasets/mta_perf/Performance_MNR.xml'
```

```python
with open(path) as f:
    parsed = objectify.parse(f)
root = parsed.getroot()
```

```python
data = []
skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
 "DESIRED_CHANGE", "DECIMAL_PLACES"]
for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
pref = pd.DataFrame(data)
pref.head()
```

pandas’s pandas.read_xml function turns this process into a one-line expression

```python
pref2 = pd.read_xml(path)
pref2
```
