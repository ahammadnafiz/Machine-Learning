```python
import pandas as pd
import numpy as np
```

## Apply: General split-apply-combine

```python
tips = pd.read_csv("examples/tips.csv")
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
def top(df, n = 5, column = 'tip_pct'):
    return df.sort_values(column, ascending=False)[:n]
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips_grouped = tips.groupby(['smoker', 'day'])
```

```python
tips_grouped.apply(top, n=1, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack()
```

 Inside GroupBy, when you invoke a method like describe, it is actually just a
 shortcut for:

```python
def f(group):
    return group.describe()
tips_grouped.apply(f)
```

## Suppressing the Group Keys

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

## Quantile and Bucket Analysis

```python
frame = pd.DataFrame({'data1':np.random.standard_normal(1000),
                     'data2':np.random.standard_normal(1000)})
frame
```

```python
quartiles = pd.cut(frame['data1'], 4, labels=['one', 'two', 'three', 'four'])
quartiles.head(10)
```

The Categorical object returned by cut can be passed directly to groupby. So we
 could compute a set of group statistics for the quartiles

```python
def get_stats(group):
    return pd.DataFrame({'min':group.min(), 'max':group.max(),
                        'count':group.count(), 'mean': group.mean()})
```

```python
grouped = frame.groupby(quartiles)
```

```python
grouped.apply(get_stats)
```

```python
grouped.agg(['min', 'max', 'count', 'mean'])
```

```python
quartiles_samp = pd.qcut(frame["data1"], 4, labels=False)
quartiles_samp.head()
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
```

```python
s.fillna(s.mean())
```

```python
states = ["Ohio", "New York", "Vermont", "Florida",
           "Oregon", "Nevada", "California", "Idaho"]
group_key = ["East", "East", "East", "East",
              "West", "West", "West", "West"]
```

```python
data = pd.Series(np.random.standard_normal(8), index=states)
```

```python
data
```

```python
data[["Vermont", "Nevada", "Idaho"]] = np.nan
```

```python
data
```

```python
data.groupby(group_key).size()
```

```python
data.groupby(group_key).count()
```

```python
data.groupby(group_key).mean()
```

```python
def fill_mean(group):
    return group.fillna(group.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
fill_values = {'East':0.5, 'West':-1}
def fill_func(group):
    return group.fillna(fill_values[group.name])
data.groupby(group_key).apply(fill_func)
```

## Group Weighted Average and Correlation

```python
df = pd.DataFrame({"category": ["a", "a", "a", "a",                    
"b", "b", "b", "b"],
 "data": np.random.standard_normal(8),      
"weights": np.random.uniform(size=8)})
```

```python
df
```

```python
grouped = df.groupby('category')
```

```python
def get_wavg(group):
    return np.average(group['data'], weights=group['weights'])
```

```python
grouped.apply(get_wavg)
```

```python
 close_px = pd.read_csv("examples/stock_px.csv", parse_dates=True,
                      index_col=0)
```
