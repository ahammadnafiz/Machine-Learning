```python
# Extract unique job titles
job_titles = df['job_title'].unique()

# Save job titles to a text file
with open('job_titles.txt', 'w', encoding='utf-8') as f:
    for title in job_titles:
        f.write(f"{title}\n")

print(f"Job titles have been saved to 'job_titles.txt'")
```

```python
# Load job titles from the text file using utf-8 encoding
with open('job_titles.txt', 'r', encoding='utf-8') as file:
    job_titles = file.readlines()

# Define a function to categorize titles based on more detailed keywords
def categorize_title(title):
    title = title.lower()  # Convert to lowercase for easy matching
    
    if 'intern' in title or 'internship' in title:
        return 'Internship'
    elif 'engineer' in title or 'engineering' in title:
        if 'software' in title or 'developer' in title:
            return 'Software Engineering'
        elif 'data' in title or 'ai' in title or 'ml' in title:
            return 'AI/ML Engineering'
        elif 'hardware' in title or 'electrical' in title:
            return 'Hardware Engineering'
        elif 'systems' in title:
            return 'Systems Engineering'
        elif 'telecom' in title or '5g' in title or 'rf' in title:
            return 'Telecommunications Engineering'
        else:
            return 'General Engineering'
    elif 'data scientist' in title:
        return 'Data Science'
    elif 'data analyst' in title:
        return 'Data Analyst'
    elif 'actuary' in title or 'actuarial' in title:
        return 'Actuary'
    elif 'finance' in title or 'financial' in title or 'treasury' in title:
        return 'Finance'
    elif 'consultant' in title or 'consulting' in title:
        return 'Consulting'
    elif 'manager' in title or 'director' in title or 'chief' in title or 'vp' in title:
        return 'Management'
    elif 'health' in title or 'medical' in title or 'pharmacy' in title:
        return 'Healthcare'
    elif 'marketing' in title:
        return 'Marketing'
    elif 'sales' in title or 'business development' in title:
        return 'Sales'
    elif 'law' in title or 'legal' in title or 'compliance' in title:
        return 'Legal/Compliance'
    elif 'education' in title or 'training' in title:
        return 'Education'
    elif 'research' in title or 'scientist' in title:
        return 'Research'
    elif 'hr' in title or 'human resources' in title:
        return 'Human Resources'
    elif 'operations' in title:
        return 'Operations'
    elif 'customer' in title or 'client' in title:
        return 'Customer Service'
    elif 'analytics' in title or 'analyst' in title:
        return 'Analytics'
    else:
        return 'Other'

# Categorize each job title
categories = [categorize_title(title) for title in job_titles]

# Create a DataFrame with the original and categorized titles
job_title_categories = pd.DataFrame({
    'Job Title': job_titles,
    'Category': categories
})

# Save to a new file (optional)
job_title_categories.to_csv('categorized_job_titles.csv', index=False)
```

```python
import re

# Define a function to categorize job titles
def categorize_job_title(title):
    title = title.lower()
    if re.search(r'software|firmware|developer', title):
        return 'Software Engineering'
    elif re.search(r'data science|analyst|data scientist|big data', title):
        return 'Data Science & Analytics'
    elif re.search(r'machine learning|ml|ai|artificial intelligence', title):
        return 'Machine Learning / AI'
    elif re.search(r'firmware|embedded', title):
        return 'Firmware & Embedded Systems'
    elif re.search(r'5g|6g|telecom|modem|rf|cellular', title):
        return 'Telecom & 5G/6G'
    elif re.search(r'actuary|actuarial', title):
        return 'Actuarial Science'
    elif re.search(r'system engineer|systems engineering', title):
        return 'System Engineering'
    elif re.search(r'algorithm|signal processing', title):
        return 'Algorithm Engineering'
    elif re.search(r'rf|radio frequency', title):
        return 'RF Engineering'
    elif re.search(r'director|manager|chief|lead', title):
        return 'Management & Leadership'
    elif re.search(r'consultant|consulting', title):
        return 'Consulting'
    elif re.search(r'intern|internship|trainee', title):
        return 'Internships'
    elif re.search(r'finance|treasury', title):
        return 'Finance & Treasury'
    elif re.search(r'data engineer|data engineering', title):
        return 'Data Engineering'
    elif re.search(r'research|development|r&d', title):
        return 'Research & Development'
    else:
        return 'Other'

# Apply the categorization to the job_title column
df['job_category'] = df['job_title'].apply(categorize_job_title)

# Display the new categorized dataframe and some job categories
df[['job_title', 'job_category']].head(20)
```

```python
df.head(4)
```

```python
# Function to shorten job titles by focusing on key terms
def shorten_job_title(title):
    # Remove any text after a comma, dash, or parentheses
    title = re.split(r'[,-]', title)[0]
    
    # Remove common phrases that are not essential for title length
    title = re.sub(r'intern|internship|senior|staff|principal|Sr.|global mobility|summer \d+|wi|cork|remote|hybrid', '', title, flags=re.IGNORECASE)
    
    # Remove extra spaces and unnecessary terms
    title = title.strip()
    return title

# Apply the function to shorten job titles
df['short_job_title'] = df['job_title'].apply(shorten_job_title)

# Display the shortened titles along with their categories
df[['job_title', 'short_job_title', 'job_category']].sample(20)
```
